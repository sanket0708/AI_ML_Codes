```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['director'].fillna('Unknown',inplace=True)
df['cast'].fillna('Unknown',inplace=True)
df['country'].fillna('Unknown',inplace=True)
```

```python
df.head()
```

```python
df['rating'].fillna("Unknown", inplace=True)

# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year
df['year_added'] = df['date_added'].dt.year
```

```python
df.head()
```

```python
df['year_added'] = df['year_added'].astype('Int64')
```

```python
df.head()
```

```python
df['type'].value_counts()
```

```python
#top 10 countries with most shows
df['country'].value_counts().head(10)
```

```python
df['release_year'].value_counts().sort_index().plot(kind='bar',figsize=(15,5))
```

```python
from textwrap import fill
#Group by release_year and type

content_trend = df.groupby(['release_year','type']).size().unstack(fill_value=0)
content_trend.tail(10)
```

```python
import matplotlib.pyplot as plt
content_trend.plot(kind='line',figsize=(15,5))
plt.title("Movies vs Shows added each year")
plt.xlabel("Release year")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
sns.countplot(x='year_added',hue='type',data=df)
```

```python
df['month_added'] = df['date_added'].dt.month
df['month_added'] = df['month_added'].astype('Int64')
```

```python
df.head()
```

```python
month_counts = df['month_added'].value_counts().sort_index()
month_counts
```

```python
import calendar

plt.figure(figsize=(10,6))
plt.bar(month_counts.index, month_counts.values, color="skyblue")
plt.xticks(month_counts.index, [calendar.month_abbr[m] for m in month_counts.index])
plt.title("Content Added by Month on Netflix")
plt.xlabel("Month")
plt.ylabel("Number of Titles Added")
plt.show()
```

```python
#top 10 countries with most shows
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(10,6))
top_countries.plot(kind='bar', color="coral")
plt.title("Top 10 Countries with Most Netflix Titles")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Split multiple countries into separate rows
country_df = df['country'].dropna().str.split(',').explode().str.strip()

# Count top 10 countries
top_countries2 = country_df.value_counts().head(10)
top_countries2
```

```python
plt.figure(figsize=(10,6))
top_countries2.plot(kind='bar', color="coral")
plt.title("Top 10 Countries with Most Netflix Titles")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

```python
genres = df['listed_in'].str.split(',').explode()
genres.value_counts().head(10)
```

```python
top_genres = genres.value_counts().head(10)
top_genres.plot(kind='bar',figsize=(15,5))
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.show()
```
